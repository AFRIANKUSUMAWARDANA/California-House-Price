#### **1.5 | Metric Evaluation**

Evaluasi metrik yang akan digunakan dalam proyek ini adalah **Root Mean Squared Error (RMSE)**, **Mean Absolute Error (MAE)**, **Mean Absolute Percentage Error (MAPE)**, dan **R-squared (R²)**.

1. **RMSE (Root Mean Squared Error):**  
   RMSE adalah akar dari rata-rata kuadrat kesalahan prediksi. RMSE memberikan penalti lebih besar untuk kesalahan prediksi yang besar, sehingga sensitif terhadap outliers. Metrik ini penting untuk memberikan gambaran tentang seberapa besar kesalahan prediksi dalam unit dolar.

2. **MAE (Mean Absolute Error):**  
   MAE adalah rata-rata dari nilai absolut kesalahan prediksi. MAE memberikan interpretasi langsung dalam unit dolar, sehingga mudah dipahami oleh stakeholder. Metrik ini relevan untuk memahami rata-rata deviasi absolut dari prediksi terhadap nilai aktual.

3. **MAPE (Mean Absolute Percentage Error):**  
   MAPE adalah rata-rata dari persentase kesalahan prediksi terhadap nilai aktual. MAPE memberikan gambaran tingkat kesalahan relatif, yang sangat berguna untuk membandingkan performa model di berbagai rentang nilai harga. MAPE memungkinkan kita mengukur performa model secara persentase, yang dapat membantu dalam konteks pengambilan keputusan bisnis.

4. **R-squared (R²):**  
   R² adalah proporsi variansi dalam variabel dependen (harga rumah) yang dapat dijelaskan oleh variabel independen (fitur-fitur dalam model). Nilai R² berkisar antara 0 dan 1; semakin mendekati 1, semakin baik model menjelaskan variansi data. Metrik ini membantu menilai seberapa baik model dapat menjelaskan data secara keseluruhan.


Semakin kecil nilai RMSE, MAE, dan MAPE yang dihasilkan, berarti model semakin akurat dalam memprediksi harga rumah sesuai dengan fitur yang digunakan. Nilai RMSE dan MAE memberikan gambaran tentang seberapa besar kesalahan prediksi dalam unit dolar AS, sedangkan nilai MAPE membantu menilai tingkat kesalahan dalam persentase yang lebih mudah dibandingkan antar-rentang nilai harga.

Nilai **R-squared** digunakan untuk mengetahui seberapa baik model dapat merepresentasikan variansi keseluruhan data. Semakin mendekati 1, maka semakin fit model terhadap data observasi, menunjukkan bahwa model dapat dipercaya dalam konteks prediksi harga rumah.

> Dengan menggunakan metrik evaluasi ini, kita dapat mengukur performa model secara komprehensif dan memastikan bahwa model yang dikembangkan tidak hanya akurat secara statistik tetapi juga memberikan nilai praktis bagi pembeli dan penjual dalam konteks bisnis. Hal ini selaras dengan tujuan proyek untuk mencapai akurasi setinggi mungkin dan kesalahan serendah mungkin, serta mengidentifikasi faktor-faktor utama yang memengaruhi harga rumah.

## **3. Exploratory Data Analysis**
---

Exploratory Data Analysis (EDA) adalah tahap penting dalam proses analisis data yang bertujuan untuk memahami karakteristik dataset secara mendalam sebelum melakukan pemodelan. Melalui EDA, kita dapat mengidentifikasi pola, anomali, hubungan antar variabel, dan mengumpulkan wawasan awal yang akan sangat bermanfaat dalam tahap selanjutnya dari proses analisis. Tahap ini juga membantu dalam menentukan langkah-langkah preprocessing data yang diperlukan, seperti penanganan missing values, normalisasi, dan transformasi fitur.

Dalam EDA, terdapat tiga jenis analisis utama yang akan dibahas:

1. **Analisis Univariate**: Menganalisis distribusi variabel tunggal untuk memahami karakteristik seperti rentang, rata-rata, dan outlier.

2. **Analisis Bivariate**: Mengeksplorasi hubungan antara dua variabel, khususnya antara fitur dan target, untuk menemukan korelasi yang relevan.

3. **Analisis Korelasi**: Mengukur hubungan linear antar variabel untuk memahami keterkaitan yang dapat mempengaruhi hasil model, serta mengidentifikasi multikolinearitas.
4. **Analisis Geospasial Multivariat**: Visualisasi multivariat sering kali memberikan wawasan yang lebih mendalam dibandingkan visualisasi bivariat.

Dengan menggunakan ketiga jenis analisis ini, kita dapat membentuk pemahaman yang lebih lengkap dan mendalam tentang data, yang akan meningkatkan kualitas model dan akurasi dalam memprediksi hasil.

### **3.4 | Geopspatial Multivariate Data**

- Visualisasi multivariat sering kali memberikan wawasan yang lebih mendalam dibandingkan visualisasi bivariat. Dengan menambahkan elemen warna (hue) pada data scatter, kita dapat menambahkan dimensi ekstra untuk dianalisis, asalkan tidak ada terlalu banyak tumpang tindih pada warna.
- Salah satu bentuk visualisasi tersebut adalah plot geografis. Hal ini menarik untuk dilakukan guna memahami bagaimana faktor geografis memengaruhi berbagai fitur.
- Kita akan meninjau kolom `longitude` dan `latitude` dimana kedua kolom tersebut masing-masing adalah Garis Bujur & Garis Lintang.

```python
import geopandas as gpd
```

```python
# Ubah dataset menjadi GeoDataFrame untuk keperluan plotting geospasial
data_gdf = gpd.GeoDataFrame(
    df, 
    geometry=gpd.points_from_xy(df['longitude'], df['latitude'])
)

# Tentukan fitur-fitur yang akan diplot
features = ['population', 'median_income', 'housing_median_age', 'median_house_value']

# Buat subplot untuk 4 fitur
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

for ax, feature in zip(axes.flatten(), features):
    # Plot setiap fitur ke subplot
    data_gdf.plot(
        column=feature, 
        cmap='plasma', 
        legend=True, 
        ax=ax, 
        markersize=10,
        alpha=0.7
    )
    # Tambahkan judul pada setiap plot
    ax.set_title(f'{feature.replace("_", " ").capitalize()}')

# Sesuaikan tata letak dan tampilkan plot
plt.tight_layout()
plt.show()
```

- Berdasarkan hasil plot menggunakan scatter plot, dapat dilihat bahwa plot terlihat seperti peta California yang sesungguhnya karena latitude dan longitude merupakan data aktual dari distrik di California.
- Untuk variabel target `median_house_value`, dari grafik yang ditampilkan, dapat langsung terlihat beberapa pola.
```python
# Membuat fungsi untuk mengecek outliers
def find_anomalies(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    limit = iqr*1.5
    print(f'IQR: {iqr}')

    limit_bawah = q1 - limit
    limit_atas = q3 + limit
    print(f'limit_bawah: {limit_bawah}')
    print(f'limit_atas: {limit_atas}')        
```

### **5.7 | Tuning Model**
---

#### **5.7.1 | Training and Validation**

**XGBoost**

Berikut adalah daftar hyperparameter yang akan disesuaikan untuk mengoptimalkan performa model:

| **Nama Kolom**        | **Deskripsi & Penjelasan Kolom**                                                                                   | **Default** |
|------------------------|--------------------------------------------------------------------------------------------------------------------|-------------|
| **reg_lambda**         | Regularisasi L2. Semakin besar nilainya, model menjadi lebih konservatif/sederhana (mirip dengan Ridge Regression).| 1           |
| **max_depth**          | Kedalaman pohon (tree depth).                                                                                     | 6           |
| **learning_rate**      | Ukuran langkah untuk pembaruan guna mencegah overfitting (step size shrinkage).                                   | 0.3         |
| **n_estimators**       | Jumlah pohon dalam hutan.                                                                                         | 100         |
| **subsample**          | Proporsi data (baris) yang digunakan untuk membangun setiap pohon (% dari total baris pada training set).         | 1           |
| **gamma**              | Node akan di-split hanya jika splitting memberikan pengurangan positif pada loss function (min_impurity_decrease).| 0           |
| **colsample_bytree**   | Proporsi fitur (kolom) yang digunakan untuk setiap pohon (% dari total kolom pada training set).                  | 1           |
| **reg_alpha**          | Regularisasi L1. Semakin besar nilainya, model menjadi lebih konservatif/sederhana (mirip dengan Lasso Regression).| 1           |


**Tuning dengan RandomizedSearchCV**

1. **Parameter `n_iter`:**
   - Tidak ada aturan baku dalam menentukan jumlah iterasi untuk RandomizedSearchCV. Namun, pada pemodelan ini, digunakan **n_iter = 60** karena menurut artikel Sun Weiran, nilai ini cukup optimal untuk menjaga keseimbangan antara waktu komputasi dan akurasi model.

2. **Parameter `n_jobs`:**
   - Digunakan nilai **n_jobs = -1**, yang berarti semua prosesor pada komputer akan digunakan untuk mempercepat proses pencarian hyperparameter.

3. **Parameter `random_state`:**
   - Untuk menjaga konsistensi hasil, nilai **random_state = 2023** akan digunakan, seperti pada `train_test_split`.

4. **Metrik Evaluasi:**
   - Digunakan tiga metrik evaluasi utama (**RMSE**, **MAE**, dan **MAPE**) untuk mengevaluasi performa model. Namun, untuk optimasi hyperparameter, akan digunakan **`neg_mean_absolute_error`** sebagai metrik utama.

5. **Cross Validation:**
   - Menggunakan **K-Folds cross-validation** dengan **n_splits=5**, di mana data akan dibagi menjadi lima kelompok.
   - Parameter **shuffle=True** digunakan untuk mengacak data sebelum membaginya ke dalam kelompok.
   - Nilai **random_state=2023** tetap dipertahankan untuk menjaga konsistensi hasil.

**Keunggulan Pendekatan**

- **Efisiensi:** RandomizedSearchCV memungkinkan eksplorasi ruang hyperparameter yang luas dengan waktu komputasi yang lebih singkat dibandingkan GridSearchCV.
- **Regularisasi:** L1 dan L2 regularisasi yang diterapkan melalui `reg_alpha` dan `reg_lambda` membantu mencegah overfitting dan meningkatkan kemampuan generalisasi model.
- **Paralelisasi:** Penggunaan `n_jobs=-1` memastikan seluruh CPU tersedia digunakan secara maksimal untuk mempercepat proses tuning.
- **Akurasi:** Metrik evaluasi yang dipilih memastikan model dioptimalkan untuk meminimalkan error absolut dan relatif.

Dengan pendekatan ini, kita dapat memanfaatkan efisiensi RandomizedSearchCV untuk mengoptimalkan performa XGBoost, meningkatkan akurasi, dan menjaga waktu komputasi tetap efisien.

```python
# Import Libraries
from sklearn.model_selection import RandomizedSearchCV
```

```python
# List hyperparameter XGBoost
reg_lambda = list(np.logspace(-3, 1, 10))
max_depth = list(np.arange(1, 11))
learning_rate = list(np.arange(1, 10)/100)
n_estimators = list(np.arange(100, 201))
subsample = list(np.arange(2, 10)/10)
gamma = list(np.arange(1, 11))
colsample_bytree = list(np.arange(1, 10)/10)
reg_alpha = list(np.logspace(-3, 1, 10))

# Hyperparam space XGboost
hyperparam_space_xgb = {
    'model__max_depth': max_depth, 
    'model__learning_rate': learning_rate,
    'model__n_estimators': n_estimators,
    'model__subsample': subsample,
    'model__gamma': gamma,
    'model__colsample_bytree': colsample_bytree,
    'model__reg_alpha': reg_alpha,
    'model__reg_lambda': reg_lambda
}

# Mendefinisikan Cross Validation
crossval = KFold(n_splits=5,shuffle=True,random_state=2023)

# Benchmark model dengan hyperparameter tuning
xgb = XGBRegressor(random_state=2023, verbosity=0)

# Membuat algorithm chains
estimator_xgb = Pipeline([
    ('preprocessing',transformer),
    # ('scaler',scaler),
    ('model',xgb)
])

# Hyperparameter tuning
random_xgb = RandomizedSearchCV(
    estimator_xgb, 
    param_distributions = hyperparam_space_xgb,
    n_iter = 60,
    cv = crossval, 
    scoring = ['neg_root_mean_squared_error', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'], 
    n_jobs = -1,
    refit = 'neg_mean_absolute_error',
    random_state = 2023 
)

random_xgb.fit(X_train,y_train)
```

Setelah melakukan fitting model yang telah dituning pada **X_train** dan **y_train**, kita akan menampilkan hasil tuning dalam bentuk dataframe. Dataframe ini akan diurutkan berdasarkan nilai **RMSE**, **MAE**, dan **MAPE** untuk memudahkan analisis performa model.
